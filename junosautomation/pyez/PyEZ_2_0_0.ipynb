# Console connection using PyEZ

1. Telnet Connection
   * Over managment ip
   * Over Console Server ip/port
2. Serial Connection
3. Examples includes
   * Facts Gathering
   * cli calls
   * rpc calls
   * Load configuration over Serial connection
   * Install image over Serial connection

### Load config over Serial connection

# Structured Resource

### Structured resource enables user to define a resource in a configuration yml file using tables and views. 
### The table/view definition can be customized as per user needs. 
### Configuration yml file is used by PyEZ framework to dynamically create configuration classes which           allows user to programmatically configure and manage JUNOS device.

````python
from jnpr.junos import Device
from jnpr.junos.factory.factory_loader import FactoryLoader
import yaml

dev = Device('xxx', user='xxx', password='xxx')
dev.open()

yaml_data = \
    """---
    UserTable:
      set: system/login/user
      key-field:
        user: name
      view: userView
    userView:
      groups:
        auth: authentication
      fields:
        user: name
        uid: uid
        class_name: class
        uidgroup: uid
        fullgroup: full-name
      fields_auth:
        password: encrypted-password
      """
globals().update(FactoryLoader().load(yaml.load(yaml_data)))

# Create object of UserTable
gt = UserTable(dev)

# Set field values
gt.user = 'TestUser_1'
gt.uid = 1346
gt.class_name = 'superuser'
gt.password = 'xxxxx'

# Create xml to be loaded
gt.append()

# print xml to be loaded
print (etree.tounicode(intf.get_table_xml(), pretty_print=True))


# lock->commit->unlock configuration
gt.set()

# Read data from device (using existing get() method)
gt.get()
for item in gt:
    print 'name:', item.user
    print 'uid:', item.uid
````

````python
from jnpr.junos import Device
from jnpr.junos.factory.factory_loader import FactoryLoader
import yaml

dev = Device('xxx', user='xxx', password='xxx')
dev.open()

yaml_data = \
    """---
  InterfaceTable:
    set: interfaces/interface
    key-field:
      - name
      - unit_name
    view: InterfaceView

  InterfaceView:
    groups:
      unit: unit
      bridge: unit/family/bridge
    fields:
      name: name
      native_vlan: native-vlan-id
    fields_unit:
      unit_name  : { 'name' : { 'default' : 0 } }
      desc       : description
    fields_bridge:
      mode: interface-mode
      vlan_list: vlan-id-list
   """

globals().update(FactoryLoader().load(yaml.load(yaml_data)))

intf = InterfaceTable(dev)

intf.name = "ge-0/0/1"
intf.mode = "trunk"
intf.vlan_list = [510, 520, 530]
intf.native_vlan = 510
intf.desc = "l2-interface created"

# append record
intf.append()

print (etree.tounicode(intf.get_table_xml(), pretty_print=True))

intf.lock()
intf.load()

if intf.commit_check():
    intf.commit(sync=True)
else:
    intf.rollback()

intf.unlock()

int = intf.get()
for item in int:
    print('name: ', item.name)
    print('unit_name: ', item.unit_name)
    print('vlan_list: ', item.vlan_list)
    print('mode: ', item.mode)
    print('native_vlan: ', item.native_vlan)
    print('description: ', item.desc)

dev.close()
````

# Multi RPC error
## If the config being loaded got more then 1 error/warning, PyEZ will give all error details to the user

````python
from jnpr.junos import Device
from jnpr.junos.utils.config import Config

dev = Device(host='xx.xxx.xxx.xx', user='xxxx', password='xxxx', gather_facts=False)
dev.open()

cnf = \
"""set interfaces xe-0/0/10 ether-options 802.3ad ae10
et interfaces xe-0/0/11 ether-options 802.3ad ae11
set interfaces xe-0/0/12 ether-options 802.3ad lcap force-up
set interfaces xe-0/0/14 ether-options 802.3ad ae14"""
cu = Config(dev)
try:
    cu.load(cnf, format='set')
    cu.commit()
except Exception as ex:
    print ex.errs
    print ex
````
